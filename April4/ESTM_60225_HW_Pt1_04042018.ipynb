```python
# We'll use some made-up data 
# to exercise the many-to-many functionality.

# We'll have two datasets.  
# The first will hold info on what cities team members live in.
# The second will hold info on what ammenities those cities have.
# We will use these to determine what ammenities are available to each team member.

team_members_cities = pd.DataFrame(
    [
        {'member': 'Mike', 'city': 'Boise'},
        {'member': 'Vanjay', 'city': 'La Habra'},
        {'member': 'Madeline', 'city': 'Aliso Viejo'},
        {'member': 'Sergio', 'city': 'Orlando'}
    ])
team_members_cities
```

```python
city_amenities = pd.DataFrame(
    {'amenities': [
        'Bogus Basin', 'Boise River', 
        'Pacific Ocean', 'Aliso Viejo Park', 'AMC Theaters',
        'Disney World', 'Universal Studios', 
        'Washington Memorial', 'Lincoln Memorial', 'Smithsonian']}, 
    index=[
        'Boise', 'Boise', 
        'Aliso Viejo', 'Aliso Viejo', 'Aliso Viejo',
        'Orlando', 'Orlando',
        'Washington DC', 'Washington DC', 'Washington DC'])
city_amenities
```

```python
# PROBLEM 11 (5 Points)
# Perform a merge between these two dataframes.
p11_results = pd.merge(team_members_cities, city_amenities, left_on='city', right_index=True, right_on='city')
p11_results
```

```python
# PROBLEM 12 (5 Points)
# Vanjay doesn't show up because his city doesn't have
# any listed amenities. Specify the correct type (not category)
# of join so that Vanjay still shows up even though 
# his city has no known amenities.
p12_results = pd.merge(team_members_cities, city_amenities, left_on='city', right_index=True, right_on='city', how='left')
p12_results
```

```python
# PROBLEM 13 (5 Points)
# We are targeting Washington DC for expansion and so we'd like
# to list out its amenities even though we don't currently have 
# any team members there. Specify the correct parameters to accomplish this.
p13_results = pd.merge(team_members_cities, city_amenities, left_on='city', right_index=True, right_on='city', how='right')
p13_results
```
