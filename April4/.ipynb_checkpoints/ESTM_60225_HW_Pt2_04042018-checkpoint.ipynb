```python
# PROBLEM 15 (5 Points)
# Define a function that will evaluate the minimum value
# of the 'Height' series of a group
# and return True if that value is greater than (>) 68, 
# otherwise return False.

# Here is the function signature.
# Add the required statements.
def greater_than_68(group):
    if group['Height'].min() > 68:
        return True
    else:
        return False
```

```python
# PROBLEM 16 (5 Points) ##################################################################################################
# Use your `greater_than_68()` function with the `filter()` method
# to drop all groups from the `players_by_position` object that don't
# pass the filter.


positions_with_all_players_over_68 = players_by_position.filter(greater_than_68)['Position'].unique()
```

```python
#check your result
#ndwbb[['Height', 'Position']].sort_values(by=['Position'])
positions_with_all_players_over_68
```

```python
# PROBLEM 17 (5 Points)
# Define a function that centers the
# 'Height' series of a `DataFrameGroupBy` object.

# If the Series being evaluated isn't the Height series, return it
# without any modification.

# IMPORTANT: You can use the `name` attribute of a Series to 
# determine the column it represents.
def center_height(series):
    if series.name == 'Height':
        return series - series.mean()
    else:
        return series
```

```python
# PROBLEM 18 (5 Points) #################################################################################################
# Use your `center_height()` function with 
# the `transform()` method to drop all groups from the 
# `players_by_class` object that don't pass the filter.
height_centered_by_group = players_by_class.transform(center_height)
```

```python
# PROBLEM 19 (5 Points) #####################################################################################################
# In the next series of problems, you will use the `apply()` method to 
# add additional columns to an existing dataframe.

# Define a function that accepts DataFrame group objects
# and defines 1 new Series/column:
    # "Height_Centered": Centers height data for each group.
    
def center_height_2(dataframe):
    dataframe['Height_Centered'] = dataframe['Height'] - dataframe['Height'].mean()
    #print(dataframe.columns.tolist())
    #print(dataframe['Position'])
    #print("height" + str(dataframe['Height'].mean()))
    #print(dataframe['Height'].mean())
    return dataframe
```

```python
# PROBLEM 20 (5 Points) #####################################################################################################
# Use the "apply()" method with your `center_height_2()` function
# to add the desired columns to your `players_by_position` DataFrame
players_by_position_with_new_columns = players_by_position.apply(center_height_2)
```
