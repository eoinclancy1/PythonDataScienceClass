```python
import numpy as np
import pandas as pd
```

```python
series_1 = pd.Series(['John','Paul','Mary','Luke','Jill'])
series_2 = pd.Series(['IN','OH','NY','NY','CA'])
series_3 = pd.Series(['jsmith@example.com','paul@gmail.com','mdoe@yahoo.com','luke@hotmail.com','jill@example.com'])

customers = pd.DataFrame({'name':series_1,'state':series_2,'email':series_3})

customers
```

```python
series_4 = pd.Series(['John','John','Paul','Mary','Mary','Mary','Luke'])
series_5 = pd.Series(['2017','2018','2018','2016','2017','2017','2016'])
series_6 = pd.Series(['item1','item1','item2','item3','item1','item4','item2'])
series_6a = pd.Series([1,2,1,3,1,4,1])

orders = pd.DataFrame({'name':series_4,'year':series_5,'item':series_6,'quantity':series_6a})

orders
```

```python
series_7 = pd.Series(['item1','item2','item3','item4','item5'])
series_8 = pd.Series(['red shirt','blue shirt','black pants','hat','purple socks'])
series_9 = pd.Series([5.50,6.50,10.00,15.00,3.50])

items = pd.DataFrame({'item':series_7,'description':series_8,'price':series_9})

items
```

```python
print(customers)
print(orders)
print(items)
```

```python
pd.merge(orders,customers)
```

```python
orders_2 = pd.merge(orders,items[['item','price']])
```

```python
orders_2
```

```python
orders_2['Total Price']=orders_2['quantity']*orders_2['price']
```

```python
orders_2
```

```python
pd.merge(orders_2,items)
```

```python
customer_orders = pd.merge(customers,orders_2,how='left')
```

```python
customers
```

```python
customer_orders.fillna(0)
```

```python
pd.merge(customer_orders,items,how='left')
```

```python
all_data = pd.merge(customer_orders,items,how='outer')
```

```python
all_data
```

```python
all_data.groupby(['name','year']).agg({'price':sum,'item':'count','quantity':sum,'Total Price':sum})
```

```python
def f_filter(x):
    return x['Total Price'].sum() > 30

all_data.groupby(['name','year']).filter(f_filter)
```

```python
def t_trans(x):
    return (x - x.sum())

trans_price = all_data['Total Price'].transform(t_trans)

all_data['trans_price']=trans_price

all_data
```
